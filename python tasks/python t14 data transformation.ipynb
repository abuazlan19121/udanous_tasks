## Advanced Data Transformation

```python
# Importing Libraries
import pandas as pd
import numpy as np
from scipy.stats import boxcox
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\HP\Downloads\practice_data.xlsx")
df
```

```python
# Box cox transformation
data = df['Sales Revenue Generated']
transformed_data, lambda_val = boxcox(data)

print(f"Optimal lambda: {lambda_val:.3f}")

# Plot before & after transformation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data, kde=True, ax=ax1, color='blue')
ax1.set_title('Original Data')
sns.histplot(transformed_data, kde=True, ax=ax2, color='green')
ax2.set_title(f'Box-Cox Transformed (λ = {lambda_val:.2f})')
plt.show()
```

```python
data1 = df['Days Present']
transformed_data1, lambda_val1 = boxcox(data1)

print(f"Optimal lambda: {lambda_val1:.3f}")

# Plot before & after transformation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data1, kde=True, ax=ax1, color='violet')
ax1.set_title('Original Data')
sns.histplot(transformed_data1, kde=True, ax=ax2, color='purple')
ax2.set_title(f'Box-Cox Transformed (λ = {lambda_val:.2f})')
plt.show()
```

```python
from scipy.stats import yeojohnson

data2 = df['Sales Revenue Generated']

transformed_data2, lambda_val2 = yeojohnson(data2)

print(f"Optimal lambda: {lambda_val2:.3f}")

# Plot before & after
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data2, kde=True, ax=ax1, color='red')
ax1.set_title('Original Data')
sns.histplot(transformed_data2, kde=True, ax=ax2, color='orange')
ax2.set_title(f'Yeo-Johnson Transformed (λ = {lambda_val2:.2f})')
plt.show()
```
