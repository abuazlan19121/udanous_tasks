```python
# importing the libraries
import numpy as np
import pandas as pd
```

```python
df.columns
```

```python
# Creating the dummies
df1 = pd.get_dummies(df, columns=['gender', 'department'])
df1.head()
```

```python
# Concatenating the dataframes
df2 = pd.concat([df1, df[['Sales Revenue Generated', 'churned']]], axis=1)
df2.head()
```

```python
# Binning the data
df['Binned_Department'] = pd.cut(df['department'], bins=3, labels=['Low', 'Medium', 'High'],include_lowest=True)
df.head()
```

```python
df['Binned_Department'] = pd.qcut(df['department'].astype('category').cat.codes, q=3, labels=['Low', 'Medium', 'High'])
df.head()
```

```python
# Mapping the data
df['Mapped_Department'] = df['department'].map({'Sales': 'Low', 'Marketing': 'Medium', 'Human Resources': 'High'})
df.head()
```
