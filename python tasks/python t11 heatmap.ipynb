# Heatmap Visualization

```python
# importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the excel file 
df = pd.read_excel(r"C:\Users\HP\Downloads\heatmap_data.xlsx")
df
```

```python
# Heatmap of task completion rate
sns.heatmap(df.pivot_table(index='department', columns='gender', values='Task Completion Rates', aggfunc='mean'), annot=True)
plt.title('Heatmap of task completion rate')
plt.show()
```

```python
df.columns
```

```python
# Heatmap of Decision Making skill Rate
sns.heatmap(df.pivot_table(index='department',columns='gender',values='Decision-Making Skills Rating',aggfunc='mean'),annot=True)
plt.title('Heatmap of Decision Making skill Rate')
plt.show()
```

```python
# Heatmap of Sales Revenue Generated
sns.heatmap(df.pivot_table(index='department',columns='gender',values='Sales Revenue Generated',aggfunc='mean'),annot=True,fmt='.2f')
plt.title('Heatmap of Sales Revenue Generated')
plt.show()
```
